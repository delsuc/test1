# tests for displaying a permutation

### first using simple lines

```python
%pylab inline
figsize(6,6)
N = 25
a = 2*pi/N
tt = arange(N)
qq = tt
X = cos(a*qq)
Y = sin(a*qq)
plot(X,Y)
plot(X,Y,'o')
plot(X[[0,N-1]],Y[[0,N-1]],'*',markersize=12)
for i in [0,N-1]: #range(N):
    text(1.15*X[i]-0.05,1.15*Y[i]-0.03,str(i+1),fontsize=12)
xticks([])
yticks([])
xlim([-1.5,1.5])
ylim([-1.5,1.5])
```

### testing with arrows
usint `quiver()`

```python
%pylab inline
figsize(6,6)
N = 6
a = 2*pi/N
tt = arange(N)
qq = random.permutation(tt)
X = cos(a*tt)
Y = sin(a*tt)
quiver(X[:-1],Y[:-1],X[1:]-X[:-1],Y[1:]-Y[:-1], scale_units='xy', angles='xy', scale=1.1)
#plot(X,Y,'o')
plot(X[[0,N-1]],Y[[0,N-1]],'*',markersize=12)
for i in range(N):
    text(1.15*X[i]-0.05,1.15*Y[i]-0.03,str(i+1),fontsize=12)
xticks([])
yticks([])
xlim([-1.5,1.5])
ylim([-1.5,1.5])
```
